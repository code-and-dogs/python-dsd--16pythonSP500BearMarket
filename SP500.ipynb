```python
import pandas as pd
import pandas_datareader.data as fin
import plotly.express as px

sp500 = fin.DataReader('^GSPC', 'yahoo', start='1980-01-01', end='2022-12-16')
sp500.to_csv('sp500.csv')
sp500df = pd.read_csv('sp500.csv', index_col='Date', parse_dates=True)
sp500df.head()
```

```python
fig = px.line(title='S&P 500 Performance')
fig.add_scatter(x=sp500df.index, y=sp500df['Close'], mode='lines')
fig.show()
```

```python
blackMonday = sp500df.loc["1987-08-17":"1987-11-30", "Close"] / sp500df.loc["1987-08-17", "Close"]
rec1990 = sp500df.loc["1990-07-19":"1990-10-08", "Close"] / sp500df.loc["1990-07-19", "Close"]
dotcom = sp500df.loc["2000-08-28":"2002-09-30", "Close"] / sp500df.loc["2000-08-28", "Close"]
worldFinCrisis = sp500df.loc["2007-10-01":"2009-03-02", "Close"] / sp500df.loc["2007-10-01", "Close"]
covid = sp500df.loc["2020-02-10":"2020-03-16", "Close"] / sp500df.loc["2020-02-10", "Close"]
today = sp500df.loc["2021-12-27":"2022-12-15", "Close"] / sp500df.loc["2021-12-27", "Close"]
today
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(15,8))
ax.plot(range(len(blackMonday)), blackMonday, label='Black Monday', color='black')
ax.plot(range(len(rec1990)), rec1990, label='1990s Recession', color='grey')
ax.plot(range(len(dotcom)), dotcom, label='Dotcom', color='brown')
ax.plot(range(len(worldFinCrisis)), worldFinCrisis, label='World Financial Crisis')
ax.plot(range(len(covid)), covid, label='Covid')
ax.plot(range(len(today)), today, label='Today', color='green')
ax.legend()
```
